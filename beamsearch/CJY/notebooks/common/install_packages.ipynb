```python
import sys
import shutil
import os
import subprocess

# 1. 설치하고자 하는 패키지 리스트를 적으세요
PACKAGES = [
    "cv2",
    "kagglehub",
    "pandas",
    "requests",
    "torch",
    "torchmetrics",
    "torchvision",
    "tqdm",
    "ultralytics",
]


def safe_install():
    has_uv = bool(shutil.which("uv"))
    has_project = os.path.exists("pyproject.toml")

    # Case A: uv 프로젝트 환경 (uv init + uv venv 상태)
    if has_uv and has_project:
        print("🚀 [uv 프로젝트 감지] uv add를 실행합니다.")
        # os.system 대신 최신 가이드인 subprocess.run을 사용합니다.
        subprocess.run(["uv", "add"] + PACKAGES, check=True)

    # Case B: uv는 깔려있으나 일반 단독 가상환경인 경우
    elif has_uv:
        print("📦 [uv 환경 감지] uv pip install을 실행합니다.")
        subprocess.run(["uv", "pip", "install"] + PACKAGES, check=True)

    # Case C: 일반 팀원 환경 (uv가 없는 경우)
    else:
        print("🐍 [일반 파이썬 환경] 현재 커널에 pip install을 실행합니다.")
        # sys.executable을 쓰면 현재 노트북이 바라보는 가상환경에 정확히 안착합니다.
        subprocess.run([sys.executable, "-m", "pip", "install"] + PACKAGES, check=True)


# 실행
safe_install()
```
